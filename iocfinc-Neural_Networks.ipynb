### Perceptron algorithm

This is when we minimize our errors by slowly fitting the formula of the line to best fit our data. By slowly manipulating the weights and the bias of the line we are slowly turning the line in a way where we try to correctly cover most of our data with minimal errors.

In pseudocode, for a point with coordinates $(p,q)$, label $y$ and prediction given by the equation $\hat{y} = step(w_{1}x_{1} + w_{2}x_{2} + b)$:

* If the point is correctly classified: $\hat{y}$ = y, do nothing
* If the point is classified as positive but the label is negative (above the line) subtract $\alpha p$, $\alpha q$ and $\alpha$ __from__ $weight_{1}$, $weight_{2}$ and $b$ respectively.
* If the point is classified as negative but the label is positive (below the line) add $\alpha p$, $\alpha q$ and $\alpha$ __from__ $weight_{1}$, $weight_{2}$ and $b$ respectively.

where $\alpha$ is the __learning rate__.

```python
def prediction(X, W, b):
    return stepFunction((np.matmul(X,W)+b)[0])
```

Here we define the __prediction__ function. The arguments for the function are $X, W$ and $b$ which are the __*Features, Weights and bias*__ respectively. Note that the complete formula for prediction would be:
$$
\hat{y} = \sigma( X\,W + b)\\

```python
def perceptronStep(X, y, W, b, learn_rate = 0.01):
    # Fill in code
    for i in range(len(X)):
        y_hat = prediction(X[i],W,b)
        if y_hat - y[i] == 1:  # Above
            W[0] -= learn_rate*(X[0])  # Weight_1
            W[1] -= learn_rate*(X[1])  # Weight_2
            b -= learn_rate  # Bias
            # Why subtract? Because the pred is above
            # To pull the line closer, we decrease the slope (weigths)
        elif y_hat - y[i] == -1:  # Below
            W[0] += learn_rate*(X[0])  # Weight_1
            W[1] += learn_rate*(X[1])  # Weight_2
            b += learn_rate  # Bias
    return W, b
```

## Error functions

Now perceptron algorithm works only if the boundary (decision plane, are they the same) is linear. Most of the time the bounds are actually curves or circles or multiple lines and this is where error functions come in.

Just to recall, in machine learning the goal of the model is to come up with the best decision plane wherein the error rates are minimized. Remember the $L_1$ and $L_2$ losses from Google Developers Machine Learning Crash Course.

This was a screen capture of the definition of a loss from the Google Crash course:



So, as stated above, the goal of machine learning is to minimize loss.  Also, take note that __not all loss are equal__. There is usually a higher loss incurred when the predicted point (model) is farther from the actual value. This is the reason why we always want to find the best fit line because we want our model to be as close to the actual as possible.

Here is where error functions come in. With the help of error functions, we are able to tweak our weights and bias so that we get a model that minimizes the loss incurred empirically. One example of this would be __Gradient descent__ where we incrementally adjust our weights and bias with a step (the gradient). The mathematical operation for this is simply the _derivative_. Remember that the smaller our gradient is, the better our results become, this has been already proven during numerical methods. The goal this time is that we find a gradient size that is not too small that it will take a long time to compute the best solution to minimize loss and at the same time not too large that it is no longer diverging. [This](https://developers.google.com/machine-learning/crash-course/reducing-loss/an-iterative-approach) is a really informative module on this topic of Error function and minimizing loss.

This error function happens in the __training__ phase of our model. The goal in Machine Learning, and by extension deep learning, is to train the model to slowly adjust its weights and bias towards minimizing the loss. The idea is that any random initial point should be enough to start the training. 

Usually, in the case of a bowl-like curve, the loss would eventually come to the bottom of the curve where the loss is at the minimum. What I mean with the bowl-like curve is the convex figure. In their case, there is always 1 minimum point.

Assuming we already have our line, we would now want to start calculating our loss/error function. Checking all of the conceivable values and permutations of the weights setting for the entire data set would be very inefficient, and not to mention very compute intensive. In this case we are going to use __gradient descent__.

The first step in gradient descent is to pick the starting value. For simplicity's sake, we can say that the location of the starting point does not matter that much, this is the reason why we can often do a random see or simply set the weight to 0.

The idea of gradient descent is that it would calculate the __gradient of the loss which is equal to the derivative(slope) of the curve__. This way the loss for the possible steps are computed, in this case the weight could be moved to the left or to the right depending on which direction the least error is achieved. In the case of multiple weights, the __gradient__ is the vector of partial derivatives with respect to the weight.
$$
\nabla\,f
$$
> So the gradient in calculus actually does have a use this time. Trivia: the name of the symbol is nabla. Thankfully, the mathematical computation of the gradient is already handled by TensorFlow. But it is actually fascinating to know that the concepts I was just writing in my notebook in college is actually having some sort of function
> The gradient of the function is the vector of partial derivatives with respect to all independent variables:
$$\nabla\,f$$
The positive gradient of the function points in the direction of the greatest increase of the function. The __negative gradient points to the greatest decrease__ of the function. Since we are actually looking for the lowest error, we want to find the negative gradient of our function: $$\\-\nabla\,f$$
In machine learning, we use gradients for gradient descent. Even if we have multiple variables that we want to minimize, all we have to do is __follow the negative gradient of the function__.

In the case of neural networks, however, this is often not the case. Some neural networks are notorious with having the egg tray model where there are multiple minimum loss points so initial seed is important.

The gradient is a vector therefore it has magnitude and direction. In the case of gradient descent algorithms, the gradient is usually multiplied by a scalar called a __learning rate__ or __step size__. 

__Hyperparameters__ are the knobs that we can change in machine learning algorithms. We usually do tweaking of hyperparameters so that the learning rate is just enough to get to the minimum point efficiently. Do note that _there is always a **Goldilocks** learning rate for every regression__.

```python
import numpy as np

# Write a function that takes as input a list of numbers, and returns
# the list of values given by the softmax function.
def softmax(L):
    den = 0
    prob = []
    for i in range(len(L)):
        den += np.exp(i)
    for i in range(len(L)):
        prob.append(np.exp(i)/den)
    return prob
```

```python
def sigmoid(x):  # For feed forward
    return 1/(1+np.exp(-x))
def sigmoid_prime(x):  # For back propagation
    return sigmoid(x)(1-sigmoid(x))
def error_formula(y,output): # For feed forward
    return -y*np.log(output) - (1-output)*np.log(1-output)
def error_term_formula(y,output):  # For back propagation
    return (y-output)*(output*(1-output))#-(y-y_hat)(sigmoid_prime(x))
```

### Training of the neural network

```python
# Neural Network hyperparameters
epochs = 1000
learnrate = 0.5

# Training function
def train_nn(features, targets, epochs, learnrate):
    
    # Use to same seed to make debugging easier
    np.random.seed(42)

    n_records, n_features = features.shape
    last_loss = None

    # Initialize weights
    weights = np.random.normal(scale=1 / n_features**.5, size=n_features)

    for e in range(epochs):
        del_w = np.zeros(weights.shape)
        for x, y in zip(features.values, targets):
            # Loop through all records, x is the input, y is the target

            # Activation of the output unit
            #   Notice we multiply the inputs and the weights here 
            #   rather than storing h as a separate variable 
            output = sigmoid(np.dot(x, weights))

            # The error, the target minus the network output
            error = error_formula(y, output)

            # The error term
            #   Notice we calulate f'(h) here instead of defining a separate
            #   sigmoid_prime function. This just makes it faster because we
            #   can re-use the result of the sigmoid function stored in
            #   the output variable
            error_term = error_term_formula(y, output)

            # The gradient descent step, the error times the gradient times the inputs
            del_w += error_term * x

        # Update the weights here. The learning rate times the 
        # change in weights, divided by the number of records to average
        weights += learnrate * del_w / n_records

        # Printing out the mean square error on the training set
        if e % (epochs / 10) == 0:
            out = sigmoid(np.dot(features, weights))
            loss = np.mean((out - targets) ** 2)
            print("Epoch:", e)
            if last_loss and last_loss < loss:
                print("Train loss: ", loss, "  WARNING - Loss Increasing")
            else:
                print("Train loss: ", loss)
            last_loss = loss
            print("=========")
    print("Finished training!")
    return weights
    
weights = train_nn(features, targets, epochs, learnrate)
```

```python
# Calculate accuracy on test data
tes_out = sigmoid(np.dot(features_test, weights))
predictions = tes_out > 0.5
accuracy = np.mean(predictions == targets_test)
print("Prediction accuracy: {:.3f}".format(accuracy))
```
